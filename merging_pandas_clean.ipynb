```python
# merging data sets
df1 = pd.DataFrame({'key' : ['b', 'b', 'a', 
                             'c', 'a', 'a', 'b'],
                   'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                   'data2': range(3)})
```

```python
from IPython.display import display
```

```python
#display(df1)
#display(df2)
# merge on key
pd.merge(df1 , df2)
```

```python
# merge with specified on
df3 = pd.DataFrame( {'lkey' : ['b', 'b', 'a', 'c',
                               'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey' : ['a', 'b', 'd'],
                   'data2' : range(3)})

# default is an inner join. hence d and c are left out
# inner join similar to an intersection
# control with how, set to left, outer, right
pd.merge(df3, df4, left_on = 'lkey', 
         right_on = 'rkey')
# try outer 

df5 = pd.DataFrame({'id' : ['a','b','c'],
                   'no' : [1, 2, 3]})
df6 = pd.DataFrame({'id' : ['a', 'd', 'b'],
                   'no' : [4, 5, 6]})

display(pd.merge(df5, df6, 
        how = "outer", left_on = 'id', 
         right_on = "id"))
display(pd.merge(df5, df6, how = "inner",
       left_on = "id", right_on = "id"))
```

```python
# many to many merges
df1 = pd.DataFrame( {'key' : ['b', 'b', 'a', 
                              'c', 'a', 'b'],
                    'data1' :  range(6)})
df2 = pd.DataFrame( {'key' : ['a', 'b', 'a', 'b',
                             'd'],
                    'data2' : range(5)})
display(df1)
display(df2)
```

```python
# many to many
# cartesian product of the rows
# three b rows in left, two in right, result is 6 rows
# joins only on distinct key values
display(pd.merge(df1, df2 , how = "left"))
display(pd.merge(df1, df2, how = "inner"))
```

```python
# Merging on Index
left1 = pd.DataFrame({'key' : ['a', 'b', 'a',
                               'a', 'b', 'c'],
                     'value' : range(6)})
right1 = pd.DataFrame({'group_val' : [3.5, 7]},
                     index = ['a', 'b'])
display(left1)
display(right1)
```

```python
# To merge on index, set boolean args to True or False
# default inner ie intersect
display(pd.merge(left1,  right1, left_on = 'key', 
         right_index = True))

# outer : union
display(pd.merge(left1,  right1, left_on = 'key', 
         right_index = True, how = "outer"))
```

```python
# hierarchically
lefth = pd.DataFrame( {'key1' : ['Ohio', 'Ohio',
                                 'Ohio',
                                'Nevada', 'Nevada'],
                      'key2' : [2000, 2001, 2002, 
                               2001, 2002],
                      'data' : np.arange(5.)})
righth = pd.DataFrame(np.arange(12).reshape(6, 2),
                     index = [['Nevada', 'Nevada',
                            'Ohio', 'Ohio', 'Ohio',
                               'Ohio'
                              ],
                             [2001, 2000, 2000,
                             2000, 2001, 2002]],
                     columns = ['event1', 'event2'])
display(lefth)
display(righth)
```

```python
# Need to merge on multiple columns 
display(pd.merge(lefth, righth, left_on = ['key1', 'key2'],
        right_index = True))
# use outer 
display(pd.merge(lefth, righth, left_on = ['key1', 'key2'],
        right_index = True, how = "outer"))
```

```python
# Concat
# binding/stacking, different words, same goal
arr = np.arange(12).reshape(3, 4)
arr
# concatenate arr
np.concatenate([arr, arr], axis = 1)
```

```python
# pandas concat
s1 = pd.Series([0, 1], index = ['a', 'b'])
s2 = pd.Series([2, 3, 4], index = ['c', 'd', 'e'])
s3 = pd.Series([5, 6], index = ['f', 'g'])
# defaults to axis 0
display(pd.concat([s1, s2, s3]))
# axis 1
display(pd.concat([s1, s2, s3], axis = 1, 
                  sort = False))
# set join to inner
display(pd.concat([s1, s2, s3], axis = 1, 
                  sort = False,
       join = "inner"))
```
